```python
#ignore
g = dm.GridArray(np.stack([np.linspace(0,1,1000)]*2).T)
```

# Colormaps

The following colormaps, and their reverse via `name + "_r"`, are available via `dama.cm`, e.g. `dama.cm.toaster`.
These are in "alpha" version and subject to change.

```python
#ignore
fig, ax = plt.subplots(len(dm.cm.cms)//2, 1, figsize=(10,10))
plt.subplots_adjust(hspace=1)

i = 0
for name, cm in dm.cm.cms.items():
    if not name.endswith('_r'):
        g.plot(cmap=cm, ax=ax[i])
        ax[i].axis('off')
        ax[i].set_title(name)
        i += 1
```
