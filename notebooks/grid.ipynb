```python
%load_ext autoreload
%autoreload 2
import numpy as np
import pandas
import millefeuille as mf
```

```python
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
```

```python
a = x**2 + y**2
```

```python
data = mf.DataStack(name='data')
```

```python
points = mf.PointLayer(name='points', data={'a':a, 'x':x, 'y':y})
```

```python
data.add_layer(points)
```

```python
grid = mf.Grid()
```

```python
grid['x'].points = np.linspace(0,10,1000)
grid['y'].points = np.linspace(0,10,1000)
```

```python
data.add_layer(mf.GridLayer(name='grid', grid=grid))
```

```python
data.translate(source='points', dest='grid', var='a', method='nearest')
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.imshow(data['grid']['a'])
```
