```python
%load_ext autoreload
%autoreload 2
import numpy as np
import pandas
import millefeuille as mf
```

```python
data = mf.DataStack()
```

```python
data['points'] = mf.PointLayer()
data['points']['x'] = np.random.rand(400) * 20 - 10
data['points']['y'] = np.random.rand(400) * 20 - 10
data['points']['a'] = data['points']['x']**2 + data['points']['y']**2
```

```python
grid = mf.Grid()
grid['x'].points = np.linspace(-10,10,1200)
grid['y'].points = np.linspace(-10,10,1000)
data['grid'] = mf.GridLayer(grid)
```

```python
data.translate(source='points', dest='grid', var='a', method='cubic')
```

```python
data['grid']['a'] = np.nan_to_num(data['grid']['a'])
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=(10,10))
data['grid'].plot(fig, ax, 'a', cmap='Spectral')
#data['points'].plot_2d(fig, ax, 'x', 'y', color='w', s=5, marker='*')
```
